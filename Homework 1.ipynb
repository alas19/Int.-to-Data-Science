## Question 1
By using __urlopen__, we take the data from _IBB Açık Veri Portalı_. __Parse__ takes the data and changes data format to useable format. __encoding='utf-8'__ lets us to see Turkish characters.

```python
with urlopen("https://data.ibb.gov.tr/dataset/b47b5391-bcca-4bb3-a575-8ece68901d5d/resource/bcbeff5d-14d7-4ec0-a211-4c4d3e96cfba/download/istanbul-ehir-hatlar-iskeleleri.xml") as veri:
    raw = parse(veri,encoding='utf-8')
```

In this step, we are looking for parts of the data with _.keys_ so that we can use the required part.

```python
data = raw['kml']['Document']['Folder']['Folder']
name = []
latitude = []
longitude = []
```

Now, we need to find station name and its latitude and longitude in **data**. 

```python
for i in range(len(data)-1):
    x = data[i]['Placemark']
    for j in range(len(x)-1):
        name.insert(len(name), x[j]['name'])
        column = ('LookAt' in x[j])
        if(column):
            column= 'LookAt'
        else:
            column= 'Camera'
        latitude.insert(len(latitude), x[j][column]['latitude'])
        longitude.insert(len(longitude), x[j][column]['longitude'])
```

We created name, latitude and longitude for generate datatable.

```python
datatable = {"Station Names": name, "Latitudes": latitude, "Longtitudes": longitude}
dt = pd.DataFrame(datatable)
dt.set_index("Station Names", inplace = True)
dt
```
